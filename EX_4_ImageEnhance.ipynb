```python
#Color Space Conversion

import cv2
import matplotlib.pyplot as plt

# Reading image
img = cv2.imread("/content/deer.jpg")

# Converting image from BGR color space to RGB color space
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Converting image from BGR color space to HSV color space
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# Converting image from BGR color space to HLS color space
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
# Converting image from BGR color space to GRAY color space
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Converting image from BGR color space to YCrCb color space
img_YCrCb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
# Converting image from BGR color space to LAB color space
img_LAB = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

# Setting the grid size
plt.figure(figsize=(15, 10))

# Plotting the original image
plt.subplot(2, 3, 1)
plt.title('Original')
plt.imshow(img_rgb)

# Plotting the RGB image
plt.subplot(2, 3, 2)
plt.title('RGB')
plt.imshow(img_rgb)

# Plotting the HSV image
plt.subplot(2, 3, 3)
plt.title('HSV')
plt.imshow(img_hsv)

# Plotting the HLS image
plt.subplot(2, 3, 4)
plt.title('HLS')
plt.imshow(img_hls)

# Plotting the Gray image
plt.subplot(2, 3, 5)
plt.title('GRAY')
plt.imshow(img_gray, cmap='gray')

# Plotting the YCrCb image
plt.subplot(2, 3, 6)
plt.title('YCrCb')
plt.imshow(img_YCrCb)

# Show plots
plt.show()
```

```python
#Histogram Equalization:

import cv2
import numpy as np

# Read the image in grayscale
img = cv2.imread('/content/mist.webp', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equ = cv2.equalizeHist(img)

# Convert images to uint8 data type
img_uint8 = np.uint8(img)
equ_uint8 = np.uint8(equ)

# Concatenate the original and equalized images side by side
res = np.hstack((img_uint8, equ_uint8))

# Write the result to an output file
cv2.imwrite('res.png', res)
```

```python
#Averaging
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import os

# Check if the file exists
file_path = '/content/opencv.png'
assert os.path.exists(file_path), "File could not be read, check the file path."

# Read the image
img = cv.imread(file_path)

# Check if the image is loaded successfully
assert img is not None, "File could not be read, check the file."

# Define the kernel for averaging
kernel = np.ones((5,5), np.float32)/25

# Apply 2D convolution with the averaging kernel
dst = cv.filter2D(img, -1, kernel)

# Plotting the original and filtered images
plt.subplot(121), plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv.cvtColor(dst, cv.COLOR_BGR2RGB)), plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Blurring
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('/content/opencv.png')

# Check if the image is loaded successfully
if img is None:
    print("File could not be read, check the file.")
else:
    # Apply blur
    blurred_img = cv2.blur(img, (15, 15))  # Adjust kernel size as needed

    # Display both images side by side
    plt.figure(figsize=(10, 5))

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    # Blurred image
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(blurred_img, cv2.COLOR_BGR2RGB))
    plt.title('Blurred Image')
    plt.axis('off')

    plt.show()

```

```python
#Gradient Filters:
import cv2 as cv
from matplotlib import pyplot as plt

# Read the image in grayscale
img = cv.imread('/content/sudoko.png', cv.IMREAD_GRAYSCALE)

# Check if the image is loaded successfully
if img is None:
    print("File could not be read, check the file.")
else:
    # Convert the image to float32 data type
    img_float32 = img.astype(np.float32)

    # Apply Laplacian, Sobel X, and Sobel Y filters
    laplacian = cv.Laplacian(img_float32, cv.CV_32F)
    sobelx = cv.Sobel(img_float32, cv.CV_32F, 1, 0, ksize=5)
    sobely = cv.Sobel(img_float32, cv.CV_32F, 0, 1, ksize=5)

    # Convert images back to uint8 for display
    laplacian_uint8 = cv.convertScaleAbs(laplacian)
    sobelx_uint8 = cv.convertScaleAbs(sobelx)
    sobely_uint8 = cv.convertScaleAbs(sobely)

    # Plotting
    plt.subplot(2, 2, 1), plt.imshow(img, cmap='gray')
    plt.title('Original'), plt.xticks([]), plt.yticks([])
    plt.subplot(2, 2, 2), plt.imshow(laplacian_uint8, cmap='gray')
    plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
    plt.subplot(2, 2, 3), plt.imshow(sobelx_uint8, cmap='gray')
    plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
    plt.subplot(2, 2, 4), plt.imshow(sobely_uint8, cmap='gray')
    plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

    plt.show()

```

```python
#Canny Edge Detection:
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('/content/Messi.jpg',0)
edges = cv2.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
